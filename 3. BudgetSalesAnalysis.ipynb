```python
# Merging Sales data with Customer data

MergedData = pd.merge(SalesData, CustomerData, on ='CustomerKey')
```

### Evaluating the missing values

```python
# Plotting a line chart to figure out the Percentage of missing values in all the fields

fig = plt.figure(figsize=(18,6))
miss_MergedData = pd.DataFrame((MergedData.isnull().sum())*100/MergedData.shape[0]).reset_index()
miss_MergedData["type"] = "MergedData"
ax = sns.pointplot("index",0,data=miss_MergedData,hue="type")
plt.xticks(rotation =90,fontsize =9)
plt.title("Percentage of Missing values in MergedData")
plt.ylabel("PERCENTAGE")
plt.xlabel("COLUMNS")
ax.set_facecolor("k")
fig.set_facecolor("lightgrey")
```

```python
# Droping the fields with more than 50% of missing values

MergedData=MergedData.drop('Color',axis=1)
```
